## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
from pathlib import Path

repo_root = Path.cwd()
for parent in [repo_root] + list(repo_root.parents):
    if (parent / "data" / "raw" / "content_refresh_anonymized.csv").exists():
        repo_root = parent
        break

csv_path = repo_root / "data" / "raw" / "content_refresh_anonymized.csv"
df = pd.read_csv(csv_path)

total_assets = len(df)
down_trend_pct = (len(df[df['trend_direction'] == 'down']) / total_assets) * 100
avg_ctr_overall = df['ctr'].mean()
avg_ctr_decaying = df[df['trend_direction'] == 'down']['ctr'].mean()

print(f"Loaded data from: {csv_path}")
print(f"Total Assets: {total_assets}")
print(f"Proportion of Assets in Active Decay ('down'): {down_trend_pct:.2f}%")
print(f"Average CTR: {avg_ctr_overall:.4f}")
print(f"Decaying Assets Average CTR: {avg_ctr_decaying:.4f}")
```
