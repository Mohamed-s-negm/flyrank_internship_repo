## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
import pandas as pd

# Load ground truth distributions to assess balance and signal presence
from pathlib import Path

repo_root = Path.cwd()
for parent in [repo_root] + list(repo_root.parents):
    if (parent / "data" / "raw" / "content_refresh_anonymized.csv").exists():
        repo_root = parent
        break

csv_path = repo_root / "data" / "raw" / "content_refresh_anonymized.csv"
df = pd.read_csv(csv_path)

print("--- Target Distribution A: Scale (impression_tier) ---")
print(df['impression_tier'].value_counts(dropna=False))

print("\n--- Target Distribution B: Momentum (trend_direction) ---")
print(df['trend_direction'].value_counts(dropna=False))
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import numpy as np

# 1. Extract baseline operational feature matrix slice
features = ["content_age_days", "days_since_last_update", "engagement_rate", "avg_position", "ctr", "word_count"]
X_clean = df[features].replace([np.inf, -np.inf], np.nan).fillna(0)

# 2. Consolidate into structural Unit of Analysis Dataframe representation
unit_of_analysis_df = pd.concat([
    df[['content_id', 'client_id']], 
    X_clean, 
    df['impression_tier'].rename('target_scale'),
    df['trend_direction'].rename('target_momentum')
], axis=1)

print(f"Structural Matrix Shape: {unit_of_analysis_df.shape}\n")
print("Unit of Analysis Dataframe Sample (Granular Node Profile):")
display(unit_of_analysis_df.head(5))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Empirical verification of threshold overlap across targets
print("--- Mean Feature Behavioral Signatures Across Scale Tiers ---")
print(unit_of_analysis_df.groupby('target_scale')[['ctr', 'engagement_rate', 'content_age_days']].mean())

print("\n--- Mean Feature Behavioral Signatures Across Momentum Vectors ---")
print(unit_of_analysis_df.groupby('target_momentum')[['ctr', 'engagement_rate', 'content_age_days']].mean())
```
